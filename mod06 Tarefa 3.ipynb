```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('b1abb33c4f.csv')
df
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Categorizar APGAR5
def categoriza_apgar5(valor):
    if valor >= 0 and valor <= 3:
        return 'asfixia grave'
    elif valor >= 4 and valor <= 5:
        return 'asfixia moderada'
    elif valor >= 6 and valor <= 7:
        return 'asfixia leve'
    elif valor >= 8 and valor <= 10:
        return 'normal'

df['APGAR5_categoria'] = df['APGAR5'].apply(categoriza_apgar5)
df['APGAR5_categoria'] 
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python

g = sns.catplot(
    data=df, 
    x='APGAR5_categoria', 
    hue='SEXO', 
    col='GESTACAO', 
    kind='count',
    height=5,
    aspect=1
)
g.set_axis_labels('APGAR5 (Categoria)', 'Contagem')
g.set_titles('Tempo de Gestação: {col_name}')
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python

tabela_cruzada = pd.crosstab(df['GESTACAO'], df['APGAR5_categoria'])
print(tabela_cruzada)
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tabela_percentual = pd.crosstab(df['GESTACAO'], df['APGAR5_categoria'], normalize='index') * 100
print(tabela_percentual)
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
tabela_cruzada.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Distribuição de APGAR5 para Cada Classe de GESTACAO')
plt.xlabel('Tempo de Gestação')
plt.ylabel('Contagem')
plt.legend(title='APGAR5')
plt.show()

# Podemos observar se há uma tendência de maior número de bebês em categorias de asfixia em gestações prematuras.
```
